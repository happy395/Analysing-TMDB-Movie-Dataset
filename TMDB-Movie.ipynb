# Data Cleaning

```python
# The movies with runtime below 30 minutes need to be dropped because the runtime is too short to be considered a movie. There 
# are even movies with runtime of 0 which is not supposed to be.

df.drop(df[df.runtime <30].index, inplace=True)
```

```python
# From the analysis carried out above, the imdb_id column has 10 missing values, it's safe to delete the entire column since 
# movie lovers who will be interested in the dataset information won't be too concerned about that particular column and it 
# doesn't also affect the quality of analysis.

# The homepage column has 7930 missing values which is a very large number, filling it with the mode of the column or any other 
# value in that column will be misleading because that is where movie lovers will go to find any available information on the 
# movie and possibly download it. Therefore, it's also safe to delete the entire column from the dataset.

def drop_column(column):
    """Drops these columns entirely from the dataframe"""
    
    df.drop([column], axis =1, inplace = True)
    
drop_column('imdb_id')
drop_column('homepage')
```

```python
# The production_companies column has 1030 missing values, while the director column has 44 missing values, dropping the 
# rows with missing values will be the best, filling the missing values with values already in the column will mislead 
# potential movie producers who will like to work with certain production companies and directors based on the success 
# of the past movies they've worked on.

df.dropna(subset=['production_companies'], how ='any', inplace=True)
df.dropna(subset=['director'], how ='any', inplace=True)
```

```python
# The tagline, keywords and overview columns have some missing values which can be filled by inputting values from 
# the same column. This is the best option as these information can still be gotten from the homepage of the movie.

def fill_missing(column):
    """Fills the missing value in these columns with values in the same column"""
    
    df[column].fillna(axis=0, method='ffill', inplace=True)

fill_missing('tagline')
fill_missing('keywords')
fill_missing('overview')
```

```python
# The missing values in the genre column can be filled with the genre that occur most in the dataset. Again, this won't affect 
# the quality of analysis and movie lovers can still verify the genre of the movie from the movie homepage.

print(df['genres'].mode())
df['genres'].fillna('Drama', axis=0, inplace=True)
```

## Exploratory Data Analysis

### The number of movies released per year; this will be investigated below to determine the number of movies that have been released per year in this dataset and to know which year has the highest and lowest released movies.

```python
# Number of Movies released per year
df.groupby(['release_year']).count()['original_title']
```

```python
# MOVIES RELEASED PER YEAR

plt.figure(figsize=(14,7))
sns.set(style = 'darkgrid')
movie_per_year = df.groupby('release_year')['original_title'].count()
movie_per_year.plot.bar(x='release_year', y='original_title')
plt.xlabel('Release Year', size = 20)
plt.ylabel('Number of Movies', size = 20)
plt.title('Movies Released Per Year', size=24)
```

### 2014  has the highest number of released movies with 627 released movies while 1969 has the lowest number of released movies with 28 released movies.

### The number of movies released per year have increased greatly from the 90's till now. This is seen in the huge difference between movies released during this time frame.

### The revenue generated each year from released movies will be investigated below using a barchart to determine which year has the highest and lowest revenue generated.

```python
# REVENUE GENERATED OVER THE YEARS

plt.figure(figsize=(14,7))
sns.set(style = 'darkgrid')
revenue_trend = df.groupby('release_year')['revenue'].sum()
revenue_trend.plot.bar(x='release_year', y='revenue')
plt.xlabel('Release Year', size = 20)
plt.ylabel('Revenue', size = 20)
plt.title('Revenue Over The Years', size=24)
```

### From the barchart above, 1966 has the lowest revenue of 84,736,689 dollars while 2015 has the highest revenue of 26,759,924,070 dollars.

### Revenue realized from movies have increased over the years, this might be due to the fact that number of movies have also increased over the years.

### Top 10 movies with highest budget will be analysed below, this is to compare it with the top 10 movies with highest revenue.

```python
# Top 10 Movies with Highest Budget
df.nlargest(10, 'budget')[['original_title', 'budget']]
```

```python
# TOP 10 MOVIES WITH HIGHEST BUDGET

plt.figure(figsize=(16,8))
sns.set(style = 'darkgrid')
movie_budget = df[['original_title', 'budget']].sort_values('budget', ascending= False).head(10)
movie_budget.plot.bar(x='original_title', y='budget')
plt.xlabel('Movie Title', size = 20)
plt.ylabel('Budget', size = 20)
plt.title('Top 10 Movies With Highest Budget', size=20)
```

### From the chart above, 'Tangled' and 'John Carter' have the same budget of 260,000,000 dollars.

### 'The Warrior's Way' has the highest budget of 425,000,000 dollars.
